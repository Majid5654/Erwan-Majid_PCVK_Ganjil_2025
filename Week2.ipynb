```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('/content/drive/MyDrive/PCVK/kitten.jpg')
```

```python
plt.imshow(img)
```

2. OpenCV reads image and saves in color channel BGR (Blue Green Red) Display pixel
value and image

```python
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB) #this line to convert BGR to RGB
plt.imshow(img3)
```

3. Displaying Grayscale image, resizing, flipping, saving the resulted image

```python
img_gray = cv.imread('/content/drive/MyDrive/PCVK/kitten.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

to display a colormap with the color ‘gray’

```python
plt.imshow(img_gray, cmap='gray')
```

display a colormap with the color magma

```python
plt.imshow(img_gray, cmap='magma')
```

RGB image resized to 512 x 1024

```python
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

RGB image is displayed in a larger size with the image position upside down

```python
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)

cv.imwrite("/content/drive/MyDrive/PCVK/mandrill2.tiff", img5)

fig = plt.figure(figsize=(10, 10))  # Adjust size as needed
ax = fig.add_subplot(111)
ax.imshow(img5)
plt.axis('off')  # Optional: hides axis ticks
plt.show()
```
