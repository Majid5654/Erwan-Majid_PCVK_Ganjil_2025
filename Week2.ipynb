5.Create 5 boxes of different sizes and colors in one image. it is recommended to use
random numbers!

```python
img_6 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
canvas = img_6.copy()

h, w, _ = canvas.shape


box_sizes = [(50, 50), (80, 60), (100, 100), (120, 80), (150, 120)]


section_width = w // 5

for i, (bw, bh) in enumerate(box_sizes):

    x1 = i * section_width + 10
    y1 = np.random.randint(0, h - bh)
    x2 = x1 + bw
    y2 = y1 + bh


    color = tuple(np.random.randint(0, 256, size=3).tolist())

    cv.rectangle(canvas, (x1, y1), (x2, y2), color, -1)

plt.imshow(canvas)
plt.title("random boxes")
plt.axis("off")
plt.show()
```

6. Show the image upside down!


```python
canvas_upside = np.flipud(canvas)

plt.imshow(canvas_upside)
plt.title("Cat Image with 5 Boxes (Upside Down)")
plt.axis("off")
plt.show()
```

7. Make rectangles and circles on the face of your photo image when you are active (not a passport photo)

```python
img4 = cv.imread('/content/drive/MyDrive/PCVK/ktm.png')
```

```python
img_ktm = cv.cvtColor(img4, cv.COLOR_BGR2RGB)

canvas = img_ktm.copy()

x1, y1, x2, y2 = 200, 190, 310, 310

# Gambar rectangle
cv.rectangle(canvas, (x1, y1), (x2, y2), (255, 0, 0), 3)

# Hitung titik tengah untuk lingkaran
center_x, center_y = (x1 + x2) // 2, (y1 + y2) // 2
radius = (y2 - y1) // 3

# Gambar circle
cv.circle(canvas, (center_x, center_y), radius, (0, 255, 0), 3)

plt.imshow(canvas)
plt.axis("off")
plt.show()
```

8.Create a rectangle in the lower left corner of channel B in the RGB color space from

```python
img_edit= cv.cvtColor(img, cv.COLOR_BGR2RGB)
R, G, B = cv.split(img_edit)
h, w = B.shape
x1, y1 = 10, h - 100   # kiri bawah
x2, y2 = 120, h - 10
filename = "kitty.png"

# Gambar kotak putih (255) di channel B saja
cv.rectangle(B, (x1, y1), (x2, y2), 255, -1)

# Gabungkan lagi ke RGB
merged = cv.merge([R, G, B])

cv.putText(
    merged,
    filename,
    (50, 50),
    cv.FONT_HERSHEY_SIMPLEX,
    1,
    (255, 0, 0),
    2,
    cv.LINE_AA
)

# Tampilkan hasil
plt.imshow(merged)
plt.title("Kotak Biru di Channel B (RGB)")
plt.axis("off")
plt.show()
```
