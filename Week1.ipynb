Import library numpy,pandas,cv2,skimage,PIL,matplotlib

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
import cv2
```

read and display the image

```python
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://www.lovemeow.com/media-library/cat-mom-kittens.jpg?id=33395269&width=800&quality=85"
]

for url in urls:
    image = io.imread(url)                                  # read image
    image = cv2.resize(image, (0,0), fx=0.5, fy=0.5)        # resize image to half size
    image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)        # convert color to RGB
    final_frame = cv2.hconcat((image, image_2))             # concatenate image
    cv2_imshow(final_frame)                                 # show image
    print('\n')
```

•	see the image file size

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image : tinggi x lebar = ",tinggi,"x" ,lebar)
cv2_imshow(image_2)
```

•	create a horizontal line in the middle of the image

```python
image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255,255,255]

final_frame = cv2.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

2.Create vertical lines and diagonal lines in the output image

```python
image_4 = image_2.copy()
for y in range(tinggi):
    image_4[y, lebar // 2] = [255, 255, 255]

final_frame1 = cv2.hconcat((image_2, image_4))
cv2_imshow(final_frame1)
```

```python
image_6 = image_2.copy()

for i in range(min(tinggi, lebar)):
    image_6[i, i] = [255, 255, 255]              # diagonal  (kiri atas ke kanan bawah)
    image_6[tinggi - i - 1, i] = [255, 255, 255] # diagonal  (kiri bawah ke kanan atas)

final_frame = cv2.hconcat((image_2, image_6))
cv2_imshow(final_frame)
```

3. Draw the horizontal white lines in the middle of image with certain length

```python
tinggi, lebar, _ = image.shape
image_line = image_2.copy()
y_center = tinggi // 2

line_length = 100

start_x = (lebar // 2) - (line_length // 2)
end_x   = (lebar // 2) + (line_length // 2)

for x in range(start_x, end_x):
    image_line[y_center, x] = [255, 255, 255]

#result
final_frame = cv2.hconcat((image_2, image_line))
cv2_imshow(final_frame)
```

4. Draw a box from a collection of white pixel in certain area in the image

```python
image_box = image_2.copy()

x_min, y_min = 150, 100   # pojok kiri atas
x_max, y_max = 250, 200   # pojok kanan bawah


for y in range(y_min, y_max):
    for x in range(x_min, x_max):
        image_box[y, x] = [255, 255, 255]

#result
final_frame = cv2.hconcat((image_2, image_box))
cv2_imshow(final_frame)
```
